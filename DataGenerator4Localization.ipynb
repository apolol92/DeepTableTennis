```python
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
import random
```

```python
balls = os.listdir("balls")
for i in range(0,len(balls)):
    balls[i] = "balls/"+str(balls[i])
backgrounds = os.listdir("backgrounds")
for i in range(0,len(backgrounds)):
    backgrounds[i] = "backgrounds/"+str(backgrounds[i])
```

```python
img_width = 160
img_height = 120
i = 0
while i <= 1000:
    imgBackNumber = random.randint(0,len(backgrounds)-1)
    imgBallNumber = random.randint(0,len(balls)-1)
    size = random.randint(5, 50)
    posX = random.randint(0, img_width-size-1)
    posY = random.randint(0, img_height-size-1)

    radius = size/2.0/img_width
    center_x = (posX+(size/2))/img_width
    center_y = (posY+(size/2))/img_height
    background = cv2.resize(cv2.imread(backgrounds[imgBackNumber],cv2.IMREAD_GRAYSCALE ), (img_width, img_height))
    overlay = cv2.resize(cv2.imread(balls[imgBallNumber],cv2.IMREAD_GRAYSCALE ), (size, size))

    #print(overlay[0][0])

    for h in range(0,overlay.shape[0]):
        for w in range(0,overlay.shape[1]):
            if(overlay.item(h,w)<254 and overlay.item(h,w)>50):
                #print("before" + str(background[posY+h][posX+w]))
                background[posY+h][posX+w] = overlay.item(h,w)
                #print("after" + str(background[posY+h][posX+w]))


    cv2.imwrite("output/"+str(float(center_x))+"_"+str(float(center_y))+"_"+str(float(radius))+".png",background)
    i+=1
```
